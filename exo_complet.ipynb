```python
def saisir_notes(seuil=10.0, max_notes=10):
    notes = []
    compteur = 0  
    while compteur < max_notes:
        entree = input(f"Entrez une note ou 'stop' ({compteur+1}/{max_notes}) : ").strip()
        if entree.lower() == "stop":
            print("Saisie interrompue par l'utilisateur.")
            break
        if entree == "":
            print("Entrée vide, merci de saisir une valeur.")
            continue  

        try:
            note = float(entree)
        except ValueError:
            print("Valeur incorrecte, merci d'entrer un nombre.")
            continue

        notes.append(note)
        compteur += 1

    else:
        print(f"Nombre maximal de notes atteint ({max_notes}).")

    if not notes:
        print("Aucune note valide saisie.")
        return notes 

    print("\n--- Résultats par étudiant ---")
    for index, note in enumerate(notes, start=1):
        statut = "Admis" if note >= seuil else "Refusé"
        print(f"Étudiant {index} : note {note} → {statut}")

    moyenne = sum(notes) / len(notes)
    if moyenne >= seuil:
        mention = "Moyenne satisfaisante"
    else:
        mention = "Moyenne insuffisante"

    print("\n--- Synthèse ---")
    print(f"Nombre de notes : {len(notes)}")
    print(f"Moyenne : {moyenne:.2f} → {mention}")

    return notes
```

```python
if __name__ == "__main__":
    saisir_notes()
```
